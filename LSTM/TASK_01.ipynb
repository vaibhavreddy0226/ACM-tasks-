```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mpi_roof.csv', encoding='latin1')
df.head()
```

```python
df.info()
```

```python
df['Date Time'] = pd.to_datetime(df['Date Time'], format='%d.%m.%Y %H:%M:%S')
```

```python
df.set_index('Date Time', inplace=True)
```

```python
aggregation_dict = {
    'p (mbar)': 'mean',
    'T (degC)': 'mean',
    'Tpot (K)': 'mean',
    'Tdew (degC)': 'mean',
    'rh (%)': 'mean',
    'VPmax (mbar)': 'mean',
    'VPact (mbar)': 'mean',
    'VPdef (mbar)': 'mean',
    'sh (g/kg)': 'mean',
    'H2OC (mmol/mol)': 'mean',
    'rho (g/m**3)': 'mean',
    'wv (m/s)': 'mean',
    'max. wv (m/s)': 'max',
    'wd (deg)': 'mean',
    'rain (mm)': 'sum',
    'raining (s)': 'sum',
    'SWDR (W/m²)': 'mean',
    'PAR (µmol/m²/s)': 'mean',
    'max. PAR (µmol/m²/s)': 'max',
    'Tlog (degC)': 'mean',
    'CO2 (ppm)': 'mean'
}
```

```python
df_hourly = df.resample('1H').agg(aggregation_dict)
```

```python
df_hourly.head()
```

```python
df_hourly.corr()
```

```python
df_hourly.drop(columns=['Tpot (K)','Tlog (degC)','max. wv (m/s)','max. PAR (µmol/m²/s)','CO2 (ppm)','VPact (mbar)','VPdef (mbar)','sh (g/kg)','H2OC (mmol/mol)','rho (g/m**3)','raining (s)'], inplace=True)
```

```python
df_hourly['hour_sin'] = np.sin(2 * np.pi * df_hourly.index.hour / 24)
df_hourly['hour_cos'] = np.cos(2 * np.pi * df_hourly.index.hour / 24)
df_hourly['day_sin'] = np.sin(2 * np.pi * df_hourly.index.dayofyear / 365)
df_hourly['day_cos'] = np.cos(2 * np.pi * df_hourly.index.dayofyear / 365)
```

```python
df_hourly.head()
```

```python
from sklearn.preprocessing import StandardScaler
import joblib

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_hourly)
joblib.dump(scaler, 'scaler.pkl')
print("scaled shape:", df_scaled.shape)
```

```python
def create_multistep_sequences(data, Input, Output, target_col=0):
    X, y = [], []
    for i in range(len(data) - (Input + Output)):
        X.append(data[i : i + Input])
        y.append(data[i + Input : i + Input + Output, target_col])
    return np.array(X), np.array(y)
```

```python
Input = 720
Output = 168

X_all, y_all = create_multistep_sequences(df_scaled,Input,Output)
print("All sequences:", X_all.shape, y_all.shape)
```

```python
total_samples = len(X_all)
train_size = int(0.7 * total_samples)
val_size = int(0.15 * total_samples)

X_train = X_all[:train_size]
y_train = y_all[:train_size]

X_val = X_all[train_size:train_size + val_size]
y_val = y_all[train_size:train_size + val_size]

X_test = X_all[train_size + val_size:]
y_test = y_all[train_size + val_size:]

print("Train:", X_train.shape, y_train.shape)
print("Val  :", X_val.shape, y_val.shape)
print("Test :", X_test.shape, y_test.shape)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader

class WeatherDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
```

```python
BATCH_SIZE = 32

train_loader = DataLoader(WeatherDataset(X_train, y_train),batch_size=BATCH_SIZE,shuffle=True)
val_loader = DataLoader(WeatherDataset(X_val, y_val),batch_size=BATCH_SIZE,shuffle=False)
test_loader = DataLoader(WeatherDataset(X_test, y_test),batch_size=BATCH_SIZE,shuffle=False)
```

```python
import torch.nn as nn

class LSTMWeatherModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers=2):
        super().__init__()
        self.lstm = nn.LSTM(
            input_size=input_size,
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        lstm_out, _ = self.lstm(x)          
        last_hidden = lstm_out[:, -1, :]    
        out = self.fc(last_hidden)          
        return out
```

```python
Input_size = X_train.shape[2]   
Hidden_size = 64
Output_size = 168

model1 = LSTMWeatherModel(
    input_size=Input_size,
    hidden_size=Hidden_size,
    output_size=Output_size,
    num_layers=2
)

print(model1)
```

```python
class deepLSTMWeatherModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers, dropout):
        super().__init__()
        self.lstm = nn.LSTM(
            input_size=input_size,
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0
        )
        self.dropout = nn.Dropout(dropout)
        self.fc = nn.Linear(hidden_size, output_size)
        
    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        last_hidden = lstm_out[:, -1, :]
        last_hidden = self.dropout(last_hidden)
        out = self.fc(last_hidden)
        return out
```

```python
Input_size = X_train.shape[2]   
Hidden_size = 128
Output_size = 168

model2 = deepLSTMWeatherModel(
    input_size=Input_size,
    hidden_size=Hidden_size,
    output_size=Output_size,
    num_layers=3,
    dropout=0.2
)

print(model2)
```

```python
criterion = nn.MSELoss()
optimizer1 = torch.optim.Adam(model1.parameters(), lr=0.001)
optimizer2 = torch.optim.Adam(model2.parameters(), lr=0.001)
```

```python
print("Training model1...")
```

```python
epochs = 20

for epoch in range(epochs):
    model1.train()
    train_loss = 0
    for X_batch, y_batch in train_loader:
        optimizer1.zero_grad()
        predictions = model1(X_batch)
        loss = criterion(predictions, y_batch)
        loss.backward()
        optimizer1.step()
        train_loss += loss.item()
    train_loss /= len(train_loader)

    model1.eval()
    val_loss = 0
    with torch.no_grad():
        for X_batch, y_batch in val_loader:
            predictions = model1(X_batch)
            loss = criterion(predictions, y_batch)
            val_loss += loss.item()
    val_loss /= len(val_loader)
    print(f"Epoch [{epoch+1}/{epochs}] | "f"Train Loss: {train_loss:.4f} | "f"Val Loss: {val_loss:.4f}")
```

```python
print("Evaluating model1...")
```

```python
model1.eval()
test_loss = 0

with torch.no_grad():
    for X_batch, y_batch in test_loader:
        predictions = model1(X_batch)
        loss = criterion(predictions, y_batch)
        test_loss += loss.item()

test_loss /= len(test_loader)
print(f"Test MSE Loss: {test_loss:.4f}")
```

```python
print("Training model2...")
```

```python
epochs = 20

for epoch in range(epochs):
    model2.train()
    train_loss = 0
    for X_batch, y_batch in train_loader:
        optimizer2.zero_grad()
        predictions = model2(X_batch)
        loss = criterion(predictions, y_batch)
        loss.backward()
        optimizer2.step()
        train_loss += loss.item()
    train_loss /= len(train_loader)

    model2.eval()
    val_loss = 0
    with torch.no_grad():
        for X_batch, y_batch in val_loader:
            predictions = model2(X_batch)
            loss = criterion(predictions, y_batch)
            val_loss += loss.item()
    val_loss /= len(val_loader)
    print(f"Epoch [{epoch+1}/{epochs}] | "f"Train Loss: {train_loss:.4f} | "f"Val Loss: {val_loss:.4f}")
```

```python
print("Evaluating model2...")
```

```python
model2.eval()
test_loss = 0

with torch.no_grad():
    for X_batch, y_batch in test_loader:
        predictions = model2(X_batch)
        loss = criterion(predictions, y_batch)
        test_loss += loss.item()

test_loss /= len(test_loader)
print(f"Test MSE Loss: {test_loss:.4f}")
```
